```python
from os import listdir
from scipy.io import loadmat
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from util import *
```

```python
dir_ = '../data/bctnet/'

###################### human
coac = loadmat(dir_ + 'Coactivation_matrix.mat')
G = nx.from_numpy_matrix(coac['Coactivation_matrix'], create_using=nx.Graph())
H = nx.from_numpy_matrix(coac['Coactivation_matrix'], create_using=nx.DiGraph())
plot_distribution(H, 'human')
plot_communities(G, 'human')

fig = plt.figure(figsize=(14,14))
ax = fig.add_subplot(111)
cax = ax.matshow(coac['Coactivation_matrix'])
fig.colorbar(cax)

###################### rest
rest = loadmat(dir_ + 'GroupAverage_rsfMRI_matrix.mat')
G = nx.from_numpy_matrix(rest['GroupAverage_rsfMRI'], create_using=nx.Graph())
H = nx.from_numpy_matrix(rest['GroupAverage_rsfMRI'], create_using=nx.DiGraph())
plot_distribution(H, 'cat')
plot_communities(G, 'cat')

fig = plt.figure(figsize=(14,14))
ax = fig.add_subplot(111)
cax = ax.matshow(rest['GroupAverage_rsfMRI'])
fig.colorbar(cax)

# ###################### cat
# cat = loadmat(dir_ + 'cat.mat')
# G = nx.from_numpy_matrix(cat['CIJall'], create_using=nx.Graph())
# H = nx.from_numpy_matrix(cat['CIJall'], create_using=nx.DiGraph())
# plot_distribution(H, 'cat')
# plot_communities(G, 'cat')

# fig = plt.figure(figsize=(14,14))
# ax = fig.add_subplot(111)
# cax = ax.matshow(cat['CIJall'])
# fig.colorbar(cax)

# plot_ordered_matrix(cat['CIJall'], 4, 'cat')

# ###################### macaque
# mac = loadmat(dir_ + 'macaque71.mat')
# G = nx.from_numpy_matrix(mac['CIJ'], create_using=nx.Graph())
# H = nx.from_numpy_matrix(mac['CIJ'], create_using=nx.DiGraph())
# plot_distribution(H, 'macaque')
# plot_communities(G, 'macaque')

# fig = plt.figure(figsize=(14,14))
# ax = fig.add_subplot(111)
# cax = ax.matshow(mac['CIJ'])
# fig.colorbar(cax)

# plot_ordered_matrix(mac['CIJ'], 4, 'macaque71')
```
